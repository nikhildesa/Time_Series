# Time series analysis
It is used to predict the future values based on the previous observed values wrt time. used in forecasting

# components
Trend - upward or downward.
seasonality - occurs during a specific season
Irregurality - random occurence.
Cyclic - follows a specific cycle (sign wave).

# when not to use time series
when y is constant over time.
when y is a function of x. eg y = 2x+5

# stationarity
time series should be stationary
constant mean
constant variance - the distance from each point to the mean should be equal.
Autocovariance - y at (t -1) and y at t-2 should not have any correlations.

# Stationarity test
Rolling statistics
ADCF test

# Arima
Ar + I + MA
Auto regressive + Integration + Moving average.

# Stationary
the data shows upward trend hence the data is not stationary.
it does not have constant mean.

# lets do stationarity test

Visual technique
Rolling statistics, Rooling mean is the moving average

```python
# https://www.portent.com/blog/analytics/rolling-averages-math-moron.htm      
rolmean = indexDataset.rolling(window = 12).mean()
rolstd = indexDataset.rolling(window= 12).std()
print(rolmean)
print(rolstd)
```

```python
mean = plt.plot(rolmean,color = 'red',label = 'Rolling mean')
std = plt.plot(rolstd,color = 'black',label = 'Rolling Std')
origin = plt.plot(indexDataset,color = 'blue',label = 'original data')
plt.legend(loc = 'best')
plt.title('Rolling mean and standard deviation')
plt.show()
```

The mean is increasing and standard deviation is increasing too. The data is not stationary

```python
# Dickey-fuller test
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(indexDataset['Passengers'],autolag = 'AIC')
dfoutput = pd.Series(dftest[0:4],index = ['Test statistic','p value','lags used','No of observation'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)'%key] = value
print(dfoutput)
```

p is very large. the stationary test fails. The data is not stationary.


```python
# Estimating the trend
# Try log on data
indexDataset_logScale = np.log(indexDataset)   # Taking log of the # passengers.
plt.plot(indexDataset_logScale)
```

```python
movingAverage = indexDataset_logScale.rolling(window = 12).mean()
movingSTD = indexDataset_logScale.rolling(window= 12).std()
plt.plot(indexDataset_logScale)
plt.plot(movingAverage,color = 'Red')
```

The data is not yet stationary it is still showing an upward trend

```python
# Try log - rolling average
# sqrt - rollinf average
datasetLogScaleMinusAverage = np.subtract(indexDataset_logScale,movingAverage)
datasetLogScaleMinusAverage.dropna(inplace = True)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationary(timeseries):
    moving_average = timeseries.rolling(window = 12).mean()
    moving_std = timeseries.rolling(window = 12).std()
    
    # plot rolling statistics
    original_data = plt.plot(timeseries,color = 'Blue',label = "dataset")
    mean = plt.plot(moving_average,color = 'Red',label = "moving average")
    std = plt.plot(moving_std,color = 'black',label = "standard deviation")
    plt.legend(loc = 'best')
    plt.title('Rolling mean , standard deviation, original data')
    plt.show()
    
    # Dickey fuller
    dftest = adfuller(timeseries,autolag = 'AIC')
    dfoutput = pd.Series(dftest[0:4],index = ['Test statistic','p value','lags used','No of observation'])
    for key,value in dftest[4].items():
        dfoutput['critical value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationary(datasetLogScaleMinusAverage)
```

# The data looks very stable now , The p value is very small 0.022

```python
# lets with expo moving average instead of simple moving average

exponentialDecayWeightedAverage = indexDataset_logScale.ewm(halflife=12,min_periods=0,adjust = True).mean()
plt.plot(indexDataset_logScale,color = 'Blue')
plt.plot(exponentialDecayWeightedAverage,color = 'Red')
plt.show()
```

```python
datasetLogScaleMinusMovingExponentialDecayAverage = np.subtract(indexDataset_logScale,exponentialDecayWeightedAverage)
test_stationary(datasetLogScaleMinusMovingExponentialDecayAverage)
```

# As dataset_log - moving_average is stationary we can proceed with dataset_log

This looks good too the data is stationary. P value is small too.

```python
datasetLogDiffShifting = indexDataset_logScale - indexDataset_logScale.shift()
plt.plot(datasetLogDiffShifting)
datasetLogDiffShifting.dropna(inplace = True)
```

```python
test_stationary(datasetLogDiffShifting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(indexDataset_logScale)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.plot(indexDataset_logScale,label = "log data")
plt.legend()
plt.show()

plt.plot(trend,label = "trend")
plt.legend()
plt.show()

plt.plot(seasonal,label = "seasonal")
plt.legend()
plt.show()

plt.plot(residual,label = "residual")
plt.legend()
plt.show()

plt.plot(indexDataset_logScale,color = "blue",label = "log data")
plt.plot(trend,color = "Red",label = "Trend")
plt.plot(seasonal,color = "Black",label = "seasonal")
plt.plot(residual,color = "green",label = "residual")
plt.legend()
plt.title("data,trend,seasonal,residual")
plt.tight_layout()
plt.show()
```

```python
# Checking whether the noise is stationary or not
decomposedLogData = residual
decomposedLogData.dropna(inplace = True)
test_stationary(decomposedLogData)
```

```python
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(datasetLogDiffShifting,nlags = 20)
lag_pacf = pacf(datasetLogDiffShifting,nlags = 20,method = 'ols')

plt.plot(lag_acf)
plt.axhline(y = 0,linestyle = '--',color = 'gray')
plt.axhline(y = -1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--',color = "gray")
plt.axhline(y = 1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--',color = "gray")
plt.tight_layout()
plt.title('Autocorrelation Function ACF graph for q value')
plt.show()

plt.plot(lag_pacf)
plt.axhline(y = 0,linestyle = '--',color = 'gray')
plt.axhline(y = -1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--',color = "gray")
plt.axhline(y = 1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle = '--',color = "gray")
plt.tight_layout()
plt.title('Partial Autocorrelation Function PACF graph for p value')
plt.show()
```

# The graph drop to the confidence level i.e centre line at around 2.

p = 2 q = 2

# Arima model

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(indexDataset_logScale,order = (2,1,2))  # (p,d,q)  try (0,d,q) and (p,d,0)
result_AR = model.fit(disp = -1)
plt.plot(datasetLogDiffShifting)
plt.plot(result_AR.fittedvalues,color = 'Red')
plt.title('RSS : %.4f'%sum((result_AR.fittedvalues-datasetLogDiffShifting['Passengers'])**2))
print('Plotting AR Model')
```

```python
predictions_ARIMA_diff = pd.Series(result_AR.fittedvalues,copy = True)
print(predictions_ARIMA_diff.head())
```

```python
# cummulative sums
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_log = pd.Series(indexDataset_logScale['Passengers'].iloc[0],index = indexDataset_logScale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value = 0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(indexDataset)
plt.plot(predictions_ARIMA)
```
